```python
import pandas as pd
import numpy as np
```

```python
beml_df = pd.read_csv("BEML.csv")
beml_df[0:5]
```

```python
glaxo_df = pd.read_csv("GLAXO.csv")
glaxo_df[0:5]
```

```python
beml_df = beml_df[['Date', 'Close']]
glaxo_df = glaxo_df[['Date', 'Close']]
```

```python
beml_df
```

```python
glaxo_df
```

```python
#for line plot visualization
import matplotlib.pyplot as plt
import seaborn as sn
plt.plot(glaxo_df.Close)
plt.xlabel('Time')
plt.ylabel('Close Price')
plt.show()
```

```python
#here in line plot serial no's are taken as x-axis but we should take date and how many years it took.
#for that we have change date into date-time format by saving as index
glaxo_df = glaxo_df.set_index(pd.DatetimeIndex(glaxo_df['Date']))
beml_df = beml_df.set_index(pd.DatetimeIndex(beml_df['Date']))
beml_df
```

```python
glaxo_df
```

```python
#for Changed line plot visualization
import matplotlib.pyplot as plt
import seaborn as sn
plt.plot(glaxo_df.Close)
plt.xlabel('Time')
plt.ylabel('Close Price')
plt.show()
```

```python
#for line plot visualization
import matplotlib.pyplot as plt
import seaborn as sn
plt.plot(beml_df.Close)
plt.xlabel('Time')
plt.ylabel('Close Price')
plt.show()
```

```python
# for daily return
glaxo_df['gain'] = glaxo_df.Close.pct_change(periods = 1)
beml_df['gain'] = beml_df.Close.pct_change(periods = 1)
beml_df
glaxo_df
```

```python
#to drop null values
glaxo_df = glaxo_df.dropna()
beml_df = beml_df.dropna()
glaxo_df
```

```python
# double checking is there any null value
glaxo_df.isnull().sum()
```

```python
#plot the returns
plt.figure(figsize = (8,6))
plt.plot(glaxo_df.gain)
plt.xlabel('Time')
plt.ylabel('gain')
plt.show()
```

```python
# from these line plot we can understand avg returns are 0
#for distribution plot
import warnings
warnings.filterwarnings('ignore')
sn.distplot(glaxo_df.gain, label = 'Glaxo')
plt.xlabel('gain')
plt.ylabel('Density')
plt.legend()
plt.show()
```

```python
import warnings
warnings.filterwarnings('ignore')
sn.distplot(beml_df.gain, label = 'BEML', color ="r")
plt.xlabel('gain')
plt.ylabel('Density')
plt.legend()
plt.show()
```

```python
#from the graphs we can see that BEML has more Variance
# more variance and s.d means more risk
# combining both plots
X = pd.DataFrame(columns = ["beml_df.gain","glaxo_df.gain"])
X["beml_df.gain"] = pd.Series(beml_df.gain) # series means columns
X["glaxo_df.gain"] = pd.Series(glaxo_df.gain)
sn.distplot(X['glaxo_df.gain'],label = 'Glaxo')
sn.distplot(X['beml_df.gain'],label = 'BEML')
plt.xlabel('gain')
plt.ylabel('Density')
plt.legend()
# whichever stock is having more base, more dispersion, more S.D, more spread then
#the stock is more riskier.
```

```python
#to round of the value to four decimal places
np.round(glaxo_df['gain'].mean(),4)
```

```python
#another way to round off values
print('Mean:', np.round(glaxo_df.gain.mean(), 4))
print('Standard Deviation: ', np.round(glaxo_df.gain.std(), 4))
```

```python
print('Mean:', np.round(beml_df.gain.mean(), 4))
print('Standard Deviation: ', np.round(beml_df.gain.std(), 4))
```

```python
# what is probability of making 2% loss or more in glaxo
#means p(x<= -2%)
stats.norm.cdf(-0.02,glaxo_df.gain.mean(),glaxo_df.gain.std())
```

```python
#what is probability of making 2% gain or more in glaxo
# p(x>= 2%)
1- stats.norm.cdf( 0.02,glaxo_df.gain.mean(),glaxo_df.gain.std())
```

```python
#what is probability of making 2% gain or more in beml
#p(x>=2%)
1- stats.norm.cdf( 0.02,beml_df.gain.mean(),beml_df.gain.std())
```
